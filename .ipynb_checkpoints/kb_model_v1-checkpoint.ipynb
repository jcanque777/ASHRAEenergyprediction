```python
def convert_combo(site_id, meter, reading):
    if site_id == 0 and meter == 0: 
        reading = reading * 0.2931
    else: 
        reading = reading
    return reading 
```

```python
#combo_df = convert_combo(combo_df)
combo_df['meter_reading'] = np.vectorize(convert_combo)(combo_df['site_id'], combo_df['meter'],
                                                       combo_df['meter_reading'])
```
